```python
import pandas as pd
import streamlit as st

# Specify the actual path to your CSV file
csv_file_path = r'C:\Users\User\Desktop\sample\October 8 to October 14\Analyze Data\merged-files.csv'

# Load the CSV file into a Pandas DataFrame
df = pd.read_csv(csv_file_path)
df = df.set_index('Customer name (customer)')

# Create a filter for columns
selected_columns = st.multiselect('Select Columns to Display', df.columns)

# Filter the DataFrame based on the selected columns
filtered_df = df[selected_columns]

# Display the filtered DataFrame
st.write(filtered_df)
```
